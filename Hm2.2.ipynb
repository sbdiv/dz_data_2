# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
language_column = table.loc[:, table.columns == "Мова програмування"]

non_missing_columns = table.loc[:, table.columns != "Мова програмування"].dropna(axis=1, how='any')

table = language_column.join(non_missing_columns)

print(table)
```

# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped_table = table.groupby('Посада')
grouped_table.head()
```

# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
new_df = grouped_table.agg({"Зарплата на місяць": ["min", "max"]})
print(new_df)
new_df.shape
```

# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(group):
    return '{:.2f}'.format(group['Зарплата на місяць'].mean())
new_df['avg'] = grouped_table.apply(fill_avg_salary)
print(new_df)
```

# Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
print(new_df['avg'].describe())
```

# Збережіть отриману таблицю в CSV файл

```python
new_df.to_csv('output.csv')
```

# Створити таблицю залежності заробітної плати від займаної посади

```python
mean_salary_by_position = table.groupby('Посада')['Зарплата на місяць'].mean()
mean_salary_by_position.plot(kind='bar', figsize=(15, 6), xlabel='Посада', ylabel='Середня зарплата', title='Середня зарплата за посадами')
plt.xticks(rotation=45)
plt.show()
```
